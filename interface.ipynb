```python
import tkinter as tk
from tkinter import simpledialog
import serial
import time
import threading

# Set up the serial communication with the Arduino
arduino = serial.Serial('COM3', 9600)  # Replace 'COM3' with the correct port for your setup
time.sleep(2)  # Give some time for the connection to be established

# Global variable to control whether the hardware is running or stopped
running = False

# Function definitions
def grab_bottle():
    grab_time = simpledialog.askinteger("Input", "Enter grab time in seconds for bottle:")
    if grab_time is not None:
        release_time = simpledialog.askinteger("Input", "Enter release time in seconds for bottle:")
        if release_time is not None:
            return grab_time, release_time, 'bottle\n'
    return None, None, None

def release_bottle():
    return 'release_bottle\n'

def grab_ball():
    grab_time = simpledialog.askinteger("Input", "Enter grab time in seconds for ball:")
    if grab_time is not None:
        release_time = simpledialog.askinteger("Input", "Enter release time in seconds for ball:")
        if release_time is not None:
            return grab_time, release_time, 'ball\n'
    return None, None, None

def release_ball():
    return 'release_ball\n'

def exercise():
    grab_time = simpledialog.askinteger("Input", "Enter grab time in seconds for exercise:")
    if grab_time is not None:
        release_time = simpledialog.askinteger("Input", "Enter release time in seconds for exercise:")
        if release_time is not None:
            return grab_time, release_time, 'exercise\n'
    return None, None, None

def stop():
    return 'stop\n'

# Dictionary to store functions and their respective stop commands
funcs = {
    'bottle': (grab_bottle, release_bottle),
    'ball': (grab_ball, release_ball),
    'exercise': (exercise, stop)
}

def send_command(command):
    arduino.write(command.encode())
    arduino.flush()

def execute_task(func):
    global running
    running = True
    grab_time, release_time, command = func()
    if grab_time is not None and release_time is not None and command is not None:
        send_command(command)
        time.sleep(grab_time)
        send_command(funcs[command.strip()][1]())
        time.sleep(release_time)

def start_tasks():
    global running
    if not running:
        running = True
        for func in funcs.values():
            if running:
                threading.Thread(target=execute_task, args=(func[0],)).start()

def stop_tasks():
    global running
    running = False
    send_command(stop())

# Create the main window
root = tk.Tk()
root.title("Design and Fabrication of AI Assisted soft prosthetic hand with flexible sensing")
root.geometry('800x600')

# Create a frame with border and blue background
main_frame = tk.Frame(root, bd=2, relief=tk.SOLID, bg='blue')
main_frame.pack(fill=tk.BOTH, expand=True)

# Create and place the title and names inside the frame
title_label = tk.Label(main_frame, text="Design and Fabrication of AI Assisted soft prosthetic hand with flexible sensing", font=("Helvetica", 14, "bold"), bg='white')
title_label.pack(pady=10)

developed_by_label = tk.Label(main_frame, text="Developed by: Habibullah, Tanveer Ahmed & Alishba Shah", font=("Helvetica", 12), bg='white')
developed_by_label.pack(pady=5)

supervised_by_label = tk.Label(main_frame, text="Supervised by: Dr Afaque Manzoor", font=("Helvetica", 12), bg='white')
supervised_by_label.pack(pady=5)

co_supervised_by_label = tk.Label(main_frame, text="Co-supervised by: Engr. Umer Abdul Majeed", font=("Helvetica", 12), bg='white')
co_supervised_by_label.pack(pady=5)

special_thanks_label = tk.Label(main_frame, text="Special Thanks to: Imtiaz Ahmed", font=("Helvetica", 12), bg='white')
special_thanks_label.pack(pady=5)

# Create and place the buttons inside the frame
button_bottle = tk.Button(main_frame, text="Grab Bottle", command=lambda: execute_task(grab_bottle), width=20, height=2)
button_bottle.pack(pady=10)

button_ball = tk.Button(main_frame, text="Grab Ball", command=lambda: execute_task(grab_ball), width=20, height=2)
button_ball.pack(pady=10)

button_exercise = tk.Button(main_frame, text="Do Exercise", command=lambda: execute_task(exercise), width=20, height=2)
button_exercise.pack(pady=10)

button_start = tk.Button(main_frame, text="Start", command=start_tasks, width=20, height=2)
button_start.pack(pady=10)

button_stop = tk.Button(main_frame, text="Stop", command=stop_tasks, width=20, height=2)
button_stop.pack(pady=10)

# Start the Tkinter event loop
root.mainloop()

# Close the serial connection when the window is closed
arduino.close()
```
